```python
import pandas as pd
import scipy as scp
from pathlib import Path
import os
import time
```

```python
path_to_data = Path.cwd().parent / "Dados"
target = path_to_data / "Dados_USP.xlsx"

if not target.exists():
    raise SystemExit("Arquivo não encontrado")

df_raw = pd.read_excel(target,skiprows=1,sheet_name="Ajustes")
df_raw.head(2)
```

```python
df = df_raw.copy()

# 1. Identificamos apenas as colunas que são do tipo texto ('object')
colunas_texto = df.select_dtypes(include=['object']).columns

# 2. Aplicamos o to_timedelta em todas essas colunas de uma só vez
df[colunas_texto] = df[colunas_texto].apply(pd.to_timedelta, errors="coerce")

print(df.dtypes)
```
